```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Aman/pride/electronics_sales_data.csv")
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# ðŸ‘‰ Q1: Total Sales Revenue
total_revenue = df['Revenue'].sum()
print("Total Sales Revenue:", total_revenue)
```

```python
# ðŸ‘‰ Q2: Monthly Trend of Total Sales Revenue
df['Month'] = df['Date'].dt.to_period('M')              # Extract month
monthly_revenue = df.groupby('Month')['Revenue'].sum().reset_index()
monthly_revenue['Month'] = monthly_revenue['Month'].dt.to_timestamp()  # Convert back to datetime
```

```python
# ðŸ‘‰ Plot: Line plot using seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_revenue, x='Month', y='Revenue', marker='*')
plt.title('Monthly Total Sales Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q5: How does average revenue per unit vary by product category?

# Calculate total revenue per product
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(5).reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_products, x='Revenue', y='Product', hue='Product', palette='viridis', legend=False)
plt.title('Top 5 Best-Selling Products by Revenue')
plt.xlabel('Total Revenue ($)')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = df[['Unit_Price', 'Units_Sold', 'Revenue']].corr()

# Plot: Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python

# Q5: Calculate Revenue per Unit and plot boxplot by Category
df['Revenue_per_Unit'] = df['Revenue'] / df['Units_Sold']

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Category', y='Revenue_per_Unit')
plt.title('Average Revenue per Unit by Product Category')
plt.ylabel('Revenue per Unit')
plt.xlabel('Category')
plt.show()
```

```python

# Q6: Plot Units_Sold over time, hue by Category
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Units_Sold', hue='Category', marker='o')
plt.title('Sales Volume Over Time by Category')
plt.ylabel('Units Sold')
plt.xlabel('Date')
plt.legend(title='Category')
plt.show()
```

```python
# Aggregate total revenue per region
region_revenue = df.groupby('Region')['Revenue'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Region', y='Revenue', hue='Region', estimator=sum, dodge=False, palette='Blues')
plt.title('Total Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Pivot table: Region vs Product, showing sum of Units_Sold
pivot = df.pivot_table(index='Region', columns='Product', values='Units_Sold', aggfunc='sum', fill_value=0)

plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Units Sold by Region and Product')
plt.xlabel('Product')
plt.ylabel('Region')
plt.tight_layout()
plt.show()

```

```python
# Extract month-year for grouping
df['YearMonth'] = df['Date'].dt.to_period('M')

# Aggregate total units sold by YearMonth
monthly_sales = df.groupby('YearMonth')['Units_Sold'].sum().reset_index()

# Convert YearMonth back to datetime for plotting
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()

plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales, x='YearMonth', y='Units_Sold', marker='o')
plt.title('Monthly Total Units Sold (Seasonal Pattern)')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Extract day of the week (Monday=0, Sunday=6)
df['DayOfWeek'] = df['Date'].dt.day_name()

plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='DayOfWeek', y='Units_Sold', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Distribution of Units Sold by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.show()
```

```python
# Calculate total revenue per salesperson and get top 3
top_salespeople = df.groupby('Salesperson')['Revenue'].sum().sort_values(ascending=False).head(3).reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=top_salespeople, y='Salesperson', x='Revenue', color='steelblue')
plt.title('Top 3 Salespeople by Total Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Salesperson')
plt.tight_layout()
plt.show()
```

```python
sns.catplot(
    data=df,
    x='Salesperson',
    y='Units_Sold',
    kind='box',
    col='Region',
    col_wrap=3,
    height=4,
    aspect=1
)
plt.subplots_adjust(top=0.85)
plt.suptitle('Variance in Units Sold per Salesperson by Region')
plt.show()
```
